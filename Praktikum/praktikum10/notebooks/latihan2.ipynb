```python
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
data = {
    "NIM": ["TI001","TI002","TI003","TI004","TI005","TI006","TI007","TI008","TI009","TI010"],
    "Actual": ["Lulus","Lulus","Lulus","Lulus","Lulus",
               "Tidak Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus"],
    "Predicted": ["Lulus","Lulus","Tidak Lulus","Tidak Lulus","Lulus",
                  "Lulus","Tidak Lulus","Tidak Lulus","Lulus","Tidak Lulus"]
}

df = pd.DataFrame(data)
df
```

```python
df['Actual'] = df['Actual'].replace({"Lulus":1, "Tidak Lulus":0})
df['Predicted'] = df['Predicted'].replace({"Lulus":1, "Tidak Lulus":0})
df
```

```python
y_true = df['Actual']
y_pred = df['Predicted']

cm = confusion_matrix(y_true, y_pred)
cm
```

```python
acc = accuracy_score(y_true, y_pred)
prec = precision_score(y_true, y_pred)
rec = recall_score(y_true, y_pred)

print("Accuracy :", acc)
print("Precision:", prec)
print("Recall   :", rec)
```
